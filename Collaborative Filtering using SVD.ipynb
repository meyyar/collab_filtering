```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.metrics import mean_squared_error
from scipy.sparse.linalg import svds
import math
import itertools
import collections


class SVDRecommenderSystem(object):
    """Represents a recommender system model using Singular Value Decomposition
    """

    def __init__(self, path_to_movielens_data, svd_rank=50):
        """Initalize the model parameters, generate the model and evaluate its performance.
        """
        self.folderpath = path_to_movielens_data
        self.svd_rank = svd_rank
        self.generateRecommenderModel()
        self.evaluation()
                
    def loadData(self):
        path_to_movies_data = self.folderpath +'/movies.csv'
        path_to_ratings_data = self.folderpath + '/ratings.csv'
        self.movies_data = pd.read_csv(path_to_movies_data, encoding='utf-8')
        self.ratings_data = pd.read_csv(path_to_ratings_data, encoding='utf-8')
        self.user_movie_ratings_df = pd.merge(self.movies_data, self.ratings_data)
    
    def train_test_split(self): 
        #split dataset into 80%:20% for training:test
        self.test_df = self.user_movie_ratings_df.sample(frac = 0.2, random_state=30)
        train_df = pd.merge(self.user_movie_ratings_df, self.test_df, how='left', on=['userId', 'movieId'], indicator=True)
        self.train_df = train_df[train_df._merge == 'left_only']

    def computeRatings (self): 
        """Compute the lower rank matrix for the given User x Movies ratings matrix 
        using Singular Value Decomposition (SVD)"""
        user_movie_ratings_matrix = self.umr_train_matrix_df.as_matrix()
        self.user_ratings_mean = np.mean(user_movie_ratings_matrix, axis = 1)
        urm_matrix_demeaned = user_movie_ratings_matrix - self.user_ratings_mean.reshape(-1, 1)
        self.U, self.sigma, self.Vt = svds(urm_matrix_demeaned, self.svd_rank)
        self.sigma = np.diag(self.sigma)
        all_user_preds = np.dot(np.dot(self.U, self.sigma), self.Vt) + self.user_ratings_mean.reshape(-1, 1)
        self.rating_predictions_df =  pd.DataFrame(all_user_preds, columns=self.umr_train_matrix_df.columns, index=self.umr_train_matrix_df.index)
      
    def generateRecommenderModel (self):
        """Load the data, split it into training set and test set. 
        Then generate User x Movies matrix of ratings. 
        Then compute the singular value decomposition of this sparse User x Movies matrix.
        """
        self.loadData()
        self.train_test_split()
        self.umr_train_matrix_df = self.generateMatrix('userId', 'movieId', 'rating_x')
        self.computeRatings() 
        
    def evaluate (self): 
        sum = 0
        num_ratings = len(self.test_df)
        for index,item in self.test_df.iterrows():
            if (item.movieId in self.rating_predictions_df.columns):
                sq_error = math.pow((item.rating - self.rating_predictions_df.iloc[item.userId-1][item.movieId]), 2)
                sum+= sq_error
            else:
                num_ratings-=1
        mean = sum/num_ratings
        return math.sqrt(mean)

    def evaluation (self):
        train_rmse = math.sqrt(mean_squared_error(self.umr_train_matrix_df, self.rating_predictions_df))
        test_rmse = self.evaluate()
        print("RMSE for model with svd rank-", self.svd_rank, " is, training error : ", train_rmse, " test error : ",test_rmse)
            
    def getMoviesSeen(self, userId, display=False): 
        movies_seen = self.user_movie_ratings_df[self.user_movie_ratings_df.userId == (userId+1)]
        movies_seen = movies_seen.sort_values('rating',ascending=False)
        if (display):
            print("Graph shows genres preferred by user-", userId+1, " :")
            self.plotGenres(movies_seen.genres)
            print("Top 20 favorite movies for user-", userId+1, " :")
            print(movies_seen[['title', 'rating']].head(20).to_string(index=False))        
        return movies_seen  
    
    def getRecommendationsForUser(self, userId):
        """List the top 20 recommendations for the given userId
        """
        recommendations = self.getRecommendations (userId)
        print("Top 20 recommendations for user-", userId, " :")
        print(recommendations[['title', 'pred_rating']].to_string(index=False))
        
    def getRecommendations (self, userId): 
        userId = userId - 1
        top_unseen_rec_df = pd.DataFrame()
        if (userId in self.user_movie_ratings_df.index):
            movies_already_seen_df = self.getMoviesSeen (userId, display=True) #set display=True to view the favorite genres of the user
            ratings_pred_for_user = self.rating_predictions_df.iloc[userId].sort_values(ascending=False)
            top_100_rec_df = pd.DataFrame(ratings_pred_for_user.reset_index())[:100]
            top_100_rec_df.columns = ['movieId', 'pred_rating']
            counter = 0
            for index,item in top_100_rec_df.iterrows():
                if (item.movieId not in movies_already_seen_df.movieId.values and counter < 20 ):
                    row = self.movies_data[self.movies_data.movieId == item.movieId]
                    row['pred_rating'] = item.pred_rating
                    top_unseen_rec_df = top_unseen_rec_df.append(row)
                    counter= counter+1
            print("Genres of the recommended movies belong to :")
            self.plotGenres(top_unseen_rec_df.genres)
        return top_unseen_rec_df
    
    def predict_new_user_ratings(self):
        """Simulating new user are time 't+1', we can predict their ratings 
        using the SVD component matrices"""
        #simulate the values from that of an existing user as a new user in time t+1
        u_k = self.umr_train_matrix_df.index[1]
        #we can calculate the ratings vetor for the new user using the singular values and items-feature matrix as below.
        print("Sample of the predicted ratings for the new user in the system at time t+1:")
        print(np.dot(np.dot(self.sigma, self.Vt), u_k)[:1])
    
    #util
    def generateMatrix(self, i, c, v):
        return self.train_df.pivot(index = i, columns = c, values = v).fillna(0)
    
    #util
    def plotGenres (self, genres_column):
        genres = genres_column.str.split('|')
        genres_dict = collections.Counter(list(itertools.chain.from_iterable(genres)))
        sorted_genres_tuples = sorted(genres_dict.items(), key=lambda x: x[1], reverse=True)
        x, y = zip(*sorted_genres_tuples)
        plt.bar(range(len(x)), y, align='center')
        plt.xticks(range(len(x)), x, rotation=90)
        plt.xticks()
        plt.show()
                          
folder_path_to_dataset = 'Dataset/Collab_fitering/ml-latest-small'
userId = 500 #8
svd_rank_param = [5, 10 ,15, 25, 50] 
num_trails = len(svd_rank_param)
recommender = [None] * num_trails
for i in range(num_trails):
    print("Trail:", i+1, ", Training SVD with Rank:", svd_rank_param[i])
    #Evaluate the recommender model performance using different values for the hyperparameter of SVD, that is, the rank.
    recommender[i] = SVDRecommenderSystem(folder_path_to_dataset, svd_rank_param[i])
    #Generate the movie recommendation based on the recommender models above.
    recommender[i].getRecommendationsForUser(userId)
    print("-"*75)



#Simulate a new user at time t+1 and predict the ratings using recommender with svd rank=50:
recommender[4].predict_new_user_ratings() #consider new user being added at time t+1

```
